### Visualizing the columns and their effect on the target variable

### Graphical Analysis

```python
#plt.figure(figsize=(20,6))
ax=sns.barplot(x ='Weather_conditions', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### During the fog and cloudy days, the average time taken is high and the time is almost similar in other weather conditions

```python
plt.figure(figsize=(20,6))
ax=sns.barplot(x ='Weather_conditions', y ='Time_taken (min)', data = df,hue='Type_of_order')
for i in ax.containers:
    ax.bar_label(i)
```

#### Type of order doesnt have much effect on the time

```python

ax=sns.barplot(x ='Road_traffic_density', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### Time taken is high with increasing in traffic density (fact)

```python
ax=sns.barplot(x ='Type_of_order', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### Type of the order doesnt have any considerable effect on the time taken

```python
ax=sns.barplot(x ='Type_of_vehicle', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### Motor cycle and bicycle took more time then compared to other modes of transport

```python
ax=sns.barplot(x ='Festival', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### On festival days the time taken is 80 percent more

```python
ax=sns.barplot(x ='City', y ='Time_taken (min)', data = df)
for i in ax.containers:
    ax.bar_label(i)
```

#### Semi Urban cities took much time for delivery

```python
plt.figure(figsize=(18,8))
ax=sns.barplot(x ='City', y ='Time_taken (min)', data = df,hue = 'Type_of_vehicle')
for i in ax.containers:
    ax.bar_label(i)
```
